Not much difference between the distribution of two mba typies, so just treat equally and concat them.

```python
mba_df = pd.concat([dfs['HR_Special'], dfs['Fin_Special']], axis=0, join='outer')
mba_df
```

```python
mba_df['sl_no'].dtype
```

```python
# order by 'sl_no' ascending
MBA_df = mba_df.sort_values(by=['sl_no'], ignore_index=True)
MBA_df
```

```python
joined_df = joined_df.merge(MBA_df, how='left', left_on='sl_no', right_on='sl_no')
joined_df
```

```python
joined_df.isnull().sum()
```

```python
joined_df = joined_df.fillna(0)
joined_df.isnull().sum()
```

```python
joined_df
```

```python
joined_df["workex"] = np.where(joined_df["workex"] == "Yes", 1, 0)
joined_df
```

```python
joined_df["status"] = np.where(joined_df["status"] == "Placed", 1, 0)
joined_df["gender"] = np.where(joined_df["gender"] == "M", 1, 0)
joined_df["ssc_b"] = np.where(joined_df["ssc_b"] == "Central", 1, 0)
joined_df["hsc_b"] = np.where(joined_df["hsc_b"] == "Central", 1, 0)
joined_df["mba_t"] = np.where(joined_df["mba_t"] == "HR", 1, 0)
joined_df
```

```python
# generate binary values using get_dummies
joined_df = pd.get_dummies(joined_df, columns=["degree_t"], prefix=["degreeType_"])
joined_df
```

```python
joined_df = pd.get_dummies(joined_df, columns=["hsc_s"], prefix=["hscSpec_"])
```

```python
joined_df
```

Check whether the data obeys normal distribution (ignore the 0 values which represents 'Not placed')

```python
ax = sns.displot(joined_df.salary)
```

```python
# Use the Kolmogorov-Smirnov test
from scipy.stats import kstest

for col in joined_df.columns: # iteratre over each column
     # compare the column distribution to a normal distribution
    statistic, pval = kstest(joined_df[col], "norm")
    print(f"column = {col}") 
    print(f"p value = {pval}")
    if pval < 0.05:
        print("Probably not normal")
    else:
        print("Probaly normal")
    print("\n")
```

Obviously, not normal.

Since all features are in same scale, so no need to rescale, just add the target Y value 'salary' to the DataFrame.(use 0 value in 'salary' to reflect 'status' of 'Not placed')

```python
df_subset = joined_df[['etest_p', 'ssc_p', 'hsc_p', 'degree_p', 'mba_p', 'salary',]]
df_subset
```

```python
# use the pairplot() method to visualise pairs of variables
sns.pairplot(data = df_subset,diag_kind='hist') #diag_kind{‘auto’, ‘hist’, ‘kde’, None}
```

```python
# create a correlation matrix rounding to one decimal point
correlation_matrix = df_subset.corr().round(1)
# print a correlation heat map
sns.heatmap(data=correlation_matrix, annot=True)
```

We can see 'salary' has a positive correlation with all features, especially with 'ssc_p' and 'hsc_p' both at 0.7
Also, all features are not strongly corrleated with each other.
